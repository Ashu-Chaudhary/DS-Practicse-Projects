```python
#seprating target and input variable
df1_x=df1.drop(columns=['Happiness Score'])
y=df1[['Happiness Score']]
```

```python
#scaling the input variables for linear regression algo
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df1_x)
x=pd.DataFrame(x,columns=df1_x.columns)
```

```python
#check the skewness again
x.skew()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.25)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.25)
    reg=linear_model.LinearRegression()
    reg.fit(x_train,y_train)
    y_pred=reg.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
        
print("maxr2 score",final_r_state,"is",max_r_score)
        

```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=10,scoring="r2").mean()
```

```python
cross_val_score(linear_model.LinearRegression(),x,y,cv=10,scoring="r2")
```

```python
#max r2 score we get random state 92
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=92,test_size=0.25)

reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("r2 score is:",r2_score(y_test,y_pred))
print("RMSE is",np.sqrt(mean_squared_error(y_test,y_pred)))
```
